```python
#storing data in df variable

df = pd.read_csv('cars24data.csv')
```

```python
#checking the data

df.head(1)
```

```python
#dropping not useful columns

df.drop(columns=['Car_Name', 'RTO_code', 'Insurance', 'Insurance_Type', 'Insurance_Type.3'], inplace=True)
```

```python
#checking the data

df.head(1)
```

```python
#dropping null values

df.dropna(inplace=True)
```

```python
#Removing duplicates

df.duplicated().sum()
```

```python
#checking columns and Rows

df.shape
```

```python
#defing the Price 
# 1 price is above median 
# 0 if below

median_price = df['Price'].median()
df['Price_Class'] = (df['Price'] > median_price).astype(int)
```

```python
#checking the data

df.head(1)
```

```python
#dropping the price columns we already categorized the price

df.drop(columns=['Price'], inplace=True)
```

```python
#checking the data

df.head(1)
```

```python
# Storing  value  (X) 
# pd.get_dummies  (covertign the value to 0 and  1)
# df.drop         (use to drop the columns)
# axis=1          (Use to drop the whole column)
# drop_first=True (avoid multicollinearity)
# y = Target Variable

X = pd.get_dummies(df.drop('Price_Class', axis=1), drop_first=True) 
y = df['Price_Class'] 
```

```python
#checking the data

X.shape, y.shape
```

```python
#converting all values to 0 and 1

X = X.astype(int)
```

```python
X.head(1)
```

```python
# Train Test Split

#importing Library
from sklearn.model_selection import  train_test_split

#Splitting Data in to X and y
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# StandardScaler

#importing Libraray
from sklearn.preprocessing import StandardScaler

#Applying
sc = StandardScaler() 
X_train = sc.fit_transform(X_train)                   
X_test = sc.transform(X_test)  
```

Training The ANN Model

```python
#Library to import to Train the ANN Model

import tensorflow as tf  
import keras             
from keras.models import Sequential 
from keras.layers import Dense
```

```python
#storing in ann variable

ann = Sequential() 
```

```python
# Input layer 
ann.add(Dense(input_dim=X_train.shape[1], units=6, kernel_initializer='uniform', activation='relu')) 

# Hidden layer
ann.add(Dense(units=6, kernel_initializer='uniform', activation='relu')) 

# Output layer
ann.add(Dense(units=1, kernel_initializer='uniform', activation='sigmoid')) 
```

```python
# Compiling 

ann.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy']) 
```

```python
# Training the ANN on the Training set

ann.fit(X_train, y_train, epochs=100, batch_size=32)
```
